```python
#erwtima 3
import numpy as np
import matplotlib.pyplot as plt
#sinartisi gia ipologismo x1,x2 me gradient  descent kai sxediasmo grafikis
def gradient_descent(startingx,starttingy,learning_rate,maxiter,function,dfx,dfy):
  currx=startingx
  curry=starttingy
  iter=0
  f_arr=[]
  iter_arr=[]
  while iter<maxiter:
    prevx=currx
    prevy=curry
    currx=prevx-learning_rate*dfx(prevx,prevy)
    curry=prevy-learning_rate*dfy(prevx,prevy)
    iter=iter+1
    f_arr.append(function(currx,curry))
    iter_arr.append(iter)
    #print("\nIter:",iter,"\nvalue x1:",currx1,"\nvalue x2:",currx2) 
  plt.ylabel("times sinartisis") 
  plt.xlabel("epanalipseis") 
  plt.plot(iter_arr,f_arr)
  plt.show()
#paragwgoi prwtis sinartisis anti gia x1,x2 xrisimopoiw x,y
#https://www.wolframalpha.com/input/?i=f%28x%2C+y%29%3D%28x%E2%88%922%29%5E2%2B%28y%E2%88%923%29%5E2
f1=lambda x,y:(x-2)**2+(y-3)**2#sinartisi
df1x=lambda x,y: 2*(x-2)#paragwgos d(f(x,y)/dx1)
df1y=lambda x,y: 2*(y-3)#paragwgos d(f(x,y)/dx2)

print('\n  f1  T=10 learning rate=0.5  x1=0, x2=0')
gradient_descent(0,0,0.5,10,f1,df1x,df1y)
print('\n:  f1   1   learning rate=0.5')

print('\n  f1  T=10 learning rate=0.25  x1=0, x2=0')
gradient_descent(0,0,0.25,10,f1,df1x,df1y)
print('\n:  f1   learning rate=0.25      ')

print('\n  f1  T=10 learning rate=0.75  x1=0, x2=0')
gradient_descent(0,0,0.25,10,f1,df1x,df1y)
print('\n:  f1   learning rate=0.75      ')
print('    F1:     learning rate  0.5      ')
print(' :   learning rate       gradient descent')

#paragwgoi deuteris sinartisis anti gia x1,x2 xrisimopoiw x,y
#https://www.wolframalpha.com/input/?i=f%28x%2Cy%29%3D++%281-%28y-3%29%29%5E2%2B20%28%28x%2B3%29-%28y-3%29%5E2%29%5E2
f2=lambda x,y:(1-(y-3))**2+20*((x+3)-(y-3)**2)**2
df2x=lambda x,y:40*(x-(y-3)**2+3)
df2y=lambda x,y:2*(-40*(y-3)*(x-(y-3)**2+3)+y-4)

print('\n  f2  T=3 learning rate=0.5  x1=0, x2=0')
gradient_descent(0,0,0.5,3,f2,df2x,df2y)
print('\n: overflow error   > 3  learning rate    ')

print('\n  f2  T=5 learning rate=0.005  x1=0, x2=0')
gradient_descent(0,0,0.005,5,f2,df2x,df2y)
print('\n:   overflow error   > 5    1,2        overflow error')

print('\n  f2  T=15 learning rate=0.005   x1=-1.5,x2=3.5')
gradient_descent(-1.5,3.5,0.005,15,f2,df2x,df2y)
print('\n:   1,2       ')
print(' :      1,2               gradient descent      overflow error')

```
